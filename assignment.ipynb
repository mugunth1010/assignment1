```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url="https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=6a3dcd4b-e20a-4d52-a95a-09bf229d4721"
```

```python
response1=requests.get(url)
```
